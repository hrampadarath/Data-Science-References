```python
null_data
```

It would thus seem that way! Indeed we see the row number start is 9357. The data webpage staes there are 9358 instances, probbaly 9357? Either way these rows should be removed!

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.info()
```

Another issue with this dataset that threw me off was that most of the columns were not floats and were string representation of floats with commas as the separator! This was freaking annoying! 

```python
df.head()
```

CO(GT), C6H6(GT), T, RH and AH have this string represntation, looking at the data description(http://archive.ics.uci.edu/ml/datasets/Air+quality), these columns should be numbers and not strings/categorical data. Thus they need to e converted. To do so we can do a string replace, and convert to float. Will use chaining

```python
cols_ = ["CO(GT)","C6H6(GT)","T","AH","RH"]

for col in cols_:
    df[col] = df[col].str.replace(',','.').astype("float")
    
df.head()
```

```python
df.info()
```

```python
df.describe()
```

Looking at the stats for the columns, the minimum value for all is -200. The air tempeature cannot be -200 egrees c (Unless on the Moon, Mars, dark side of Mercury, Pluto ...). Thus the -200 is most probable a palaceholder for missing values

```python
#1. first replace the zero values with None
for col in df.columns[2:]:
    df[col].replace([-200], [np.nan], inplace=True) # always replace with np.nan
```

```python
df.isnull().sum()/df.shape[0]
```

```python
df.info()
```

### 2. Data Exploration 

NMHC(GT) is missing 90% of data! That is significant, and most probable will have to be dropped. Before moving to the next stage (imputing and model training), let's answer some questions regarding the data:
1. What is the highest CO content in June 2004? How does this compare with December 2004?
2. The most common time and weekday of the top 10 highest NO2, CO and T. 
3. The month with the highest average CO? 
4. The 5 features most correlate with temperature and absolute humidity? Are they the same? 

 

```python
df["Date"] = pd.to_datetime(df["Date"],dayfirst=True)
df.head()
```

#### 1. Highest CO in June 2006 and hows does itcompare to December?

```python
df[pd.DatetimeIndex(df["Date"]).month==6]["CO(GT)"].max()
```

```python
#Highest CO in December 2004

df[pd.DatetimeIndex(df["Date"]).month==12]["CO(GT)"].max()
```

The Max CO in December is higher than in June, is this true on avearage?

```python
df[pd.DatetimeIndex(df["Date"]).month==6]["CO(GT)"].mean()
```

```python
df[pd.DatetimeIndex(df["Date"]).month==12]["CO(GT)"].mean()
```

#### 2. The most common time and weekday of the top 10 highest NO2, CO and T. 

```python
# add new column for weekday
df["weekday"] = pd.DatetimeIndex(df["Date"]).day_name()
```

```python
#top 10 highest NO2

NO2 = df[["weekday","Time", "NO2(GT)"]].sort_values(by="NO2(GT)",ascending=False)[1:10]
NO2
```

```python
NO2.drop("NO2(GT)",axis=1).mode()
```

```python
#top 10 highest CO

CO = df[["weekday","Time", "CO(GT)"]].sort_values(by="CO(GT)",ascending=False)[1:10]
CO
```

```python
CO.drop("CO(GT)",axis=1).mode()
```

```python
#top 10 highest T

Temp = df[["weekday","Time", "T"]].sort_values(by="T",ascending=False)[1:10]
Temp
```

```python
Temp.drop("T",axis=1).mode()
```

#### 3. The month with the highest averge CO?

```python
# add a new column called month
df["Month"] = pd.DatetimeIndex(df["Date"]).month_name()
```

```python
# groupby month and select CO
df.groupby("Month")["CO(GT)"].apply(lambda x: np.mean(x))
```

Month with highest average CO is December, second October and third is November

#### 5.The 5 features most correlate with temperature and absolute humidity? Are they the same?

```python
df.corr()["T"].sort_values(ascending=False)
```

```python
df.corr()["AH"].sort_values(ascending=False)
```
